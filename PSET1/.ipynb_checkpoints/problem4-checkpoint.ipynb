```python
import numpy as np
import matplotlib.pyplot as plt
# Set a random seed for reproducibility
np.random.seed(1234)

```

```python
# params
mu = 5.0 
lambda_val = 1 / mu  
N = 10**7

# In numpy, the scale parameter is 1/lambda
X = np.random.exponential(scale=mu, size=N)

X_given = X[X > 3]

p_sim = np.mean(X_given > 9)
p_theory = np.exp(-6/mu)

print("Simulated P(X > 9 | X > 3):", p_sim)
print("Theoretical P(X > 9 | X > 3):", p_theory)
```

```python
# params
n = 100
sum_X = 496   
T_val = 28   

mu_vals = np.linspace(0.1, 10, 500)

# Full data likelihood 
L_full = mu_vals**(-n) * np.exp(-sum_X / mu_vals)

# T-likelihood 
L_T = mu_vals**(-1) * np.exp(-T_val / mu_vals) * (1 - np.exp(-T_val / mu_vals))**(n - 1)

# Normalize 
L_full_norm = L_full / np.max(L_full)
L_T_norm = L_T / np.max(L_T)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(mu_vals, L_full_norm, 'b-', lw=2, label="Full Data Likelihood")
plt.plot(mu_vals, L_T_norm, 'r--', lw=2, label="T-Likelihood")
plt.xlabel("$\\mu$", fontsize=14)
plt.ylabel("Likelihood", fontsize=14)
plt.title("Likelihood Functions for $\\mu$", fontsize=16)
plt.legend(fontsize=12)
plt.grid(False)
plt.show()
```
