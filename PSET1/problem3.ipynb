```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Given parameters
lambda_val = 7  # known
n = 10
y = np.array([1, 0, 0, 1, 1, 1, 0, 0, 1, 1])
Y = np.sum(y)  # Total number of yes responses, here Y = 6

# Define a grid for p in [0, 1]
p = np.linspace(0, 1, 200)

# Friend's (full-data) likelihood: L(p) ∝ p^Y (1-p)^(n-Y)
L_friend = p**Y * (1 - p)**(n - Y)

# Rival's likelihood: L(p) ∝ p^Y e^(-lambda*p)
L_rival = p**Y * np.exp(-lambda_val * p)

# For ease of comparison, normalize both likelihood functions by their maximum values.
L_friend_norm = L_friend / np.max(L_friend)
L_rival_norm = L_rival / np.max(L_rival)

plt.figure(figsize=(8, 6))
plt.plot(p, L_friend_norm, 'b-', lw=2, label="Friend's (Full) Likelihood: $p^6(1-p)^4$")
plt.plot(p, L_rival_norm, 'r--', lw=2, label="Rival's Likelihood: $p^6e^{-7p}$")
plt.xlabel(r'$p$', fontsize=14)
plt.ylabel('Normalized Likelihood', fontsize=14)
plt.title('Comparison of Likelihood Functions for $p$', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
